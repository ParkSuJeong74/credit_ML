### Sector 빈도수
* 빈도수가 높은 top30개만!

```python
# 상위 30개만
sector_count_top = df["Sector"].value_counts().head(30)
sector_count_top
```

```python
# 해당하는 인덱스 이름만 보기
sector_count_top.index
```

```python
# Sector 중에서 해당하는 인덱스과 같은 df만 출력
df_sector_30 = df[df["Sector"].isin(sector_count_top.index)]
df_sector_30
```

```python
# 시각화(seaborn)
plt.figure(figsize=(10,8))
sns.countplot(data=df_sector_30, y="Sector")
```

```python
# 빈도수가 높은 것부터 정렬
# 제목 설정
plt.figure(figsize=(10,8))
sns.countplot(data=df_sector_30, y="Sector",
              order=sector_count_top.index).set_title("섹터별 빈도수")
```

```python
# 색상 표
print(plt.colormaps())
```

```python
# Blues : 빈도수 낮은게 진함, Blues_r : 빈도수 높은게 진함
plt.figure(figsize=(10,8))
sns.countplot(data=df_sector_30, y="Sector", palette="Blues_r",
              order=sector_count_top.index).set_title("섹터별 빈도수")
```

## 두 변수 빈도수의 시각화
* crosstab

### 지역별 상장 종목 빈도수

```python
pd.crosstab(df["Market"], df["Region"])
```

```python
# hue : Market별로 색상 분류
plt.figure(figsize=(10,4))
plt.xticks(rotation=90) # 글씨 가로로
sns.countplot(data=df, x="Region", hue="Market")
```

```python
plt.figure(figsize=(10,4))
# 변수에 넣어주면 log 출력안됨
g=sns.countplot(data=df, y="Region", hue="Market").set_title("지역별 상장종목 빈도수")
```

### 연도별 상장 종목 빈도수

```python
market_year = pd.crosstab(df["Market"], df["ListingYear"])
market_year
```

```python
# seaborn
plt.figure(figsize=(10,4))
plt.xticks(rotation=90)
g=sns.countplot(data=df, x="ListingYear", hue="Market").set_title("연도별 상장 종목 빈도수")
```

```python
# pandas
# T : 회전
# line plot은 연속된 데이터로 가정하고 그림
g = market_year.T.plot(figsize=(10,4), title="연도별 상장 종목 빈도수")
```

```python
# 나누기
g = market_year.T.plot(figsize=(10,4), title="연도별 상장 종목 빈도수",
                        subplots=True)
```

```python
# bar plot : 범주형 데이터로 가정
g = market_year.T.plot.bar(figsize=(10,4), title="연도별 상장 종목 빈도수")
```

```python
# 컬럼 값을 int형태로 변환
market_year.columns = market_year.columns.astype(int)
market_year.columns
```

```python
g = market_year.T.plot.bar(figsize=(12,4), title="연도별 상장 종목 빈도수",
                          rot=60)
```
