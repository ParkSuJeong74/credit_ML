## Groupby와 Pivot table 시각화

* grouby -> pivot_table(느림)(쉬움) -> crosstab

### Pivot table&Groupby

```python
# pivot_table
# 기본적으로 평균
market_year=pd.pivot_table(data=df,index=["Market"] , columns="ListingYear", 
               values="Symbol", aggfunc="count", fill_value=0)
market_year
```

```python
# groupby([컬럼])[기준]
# unstack() : dataframe형태로 변환
# fillna(0) : NaN 값을 0으로 변환
market_year = df.groupby(["Market", "ListingYear"])["Symbol"].count()
market_year
```

```python
# wide form 데이터 : 시각화에 적합하진 않음
market_year.unstack().fillna(0).astype(int)
```

```python
# 각 변수가 컬럼이 되도록 변환
df_market_year=market_year.reset_index()
df_market_year=df_market_year.rename(columns={"Symbol":"count"})
df_market_year
```

### 시각화
* scatterplot, lineplot, relplot

```python
plt.figure(figsize=(10,4))
sns.scatterplot(data=df_market_year, x="ListingYear", y="count", hue="Market")
```

```python
# ci(신뢰구간, 0이면 신뢰구간 없음), n_boot(샘플링 -> 샘플 갯수가 부족하면 더)
plt.figure(figsize=(10,4))
sns.lineplot(data=df_market_year, x="ListingYear", y="count", ci=0, hue="Market")
```

```python
# subplot 가능한 relplot(lineplot과 scatterplot 호출)
sns.relplot(data=df_market_year, x="ListingYear", y="count", hue="Market",
           col="Market")
```

```python
sns.relplot(data=df_market_year, x="ListingYear", y="count", hue="Market",
           col="Market", kind="line")
```

```python
# height, aspect로 사이즈 조절
g=sns.relplot(data=df_market_year, x="ListingYear", y="count", hue="Market",
           row="Market", kind="line", height=2, aspect=3)
```
