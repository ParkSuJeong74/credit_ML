* Series.value_counts().plot()

* Series.value_counts().plot.barh()

```python
df.columns
```

### Market 빈도수

```python
# 중복 제거 : unique()
df["Market"].unique()
```

```python
# 빈도수 : value_counts()
df["Market"].value_counts()
```

```python
# 빈도수 시각화
# rot=0 : 글자 기울기없앰
df["Market"].value_counts().plot.bar(rot=0)
```

```python
df["Market"].value_counts().plot.barh()
```

```python
# seaborn에서 바로 빈도수 
sns.countplot(data=df, x="Market")
```

```python
sns.countplot(data=df, y="Market")
```

### Sector 빈도수
* 빈도수가 높은 top30개만!

```python
# 상위 30개만
sector_count_top = df["Sector"].value_counts().head(30)
sector_count_top
```

```python
# 해당하는 인덱스 이름만 보기
sector_count_top.index
```

```python
# Sector 중에서 해당하는 인덱스과 같은 df만 출력
df_sector_30 = df[df["Sector"].isin(sector_count_top.index)]
df_sector_30
```

```python
# 시각화(seaborn)
plt.figure(figsize=(10,8))
sns.countplot(data=df_sector_30, y="Sector")
```

```python
# 빈도수가 높은 것부터 정렬
# 제목 설정
plt.figure(figsize=(10,8))
sns.countplot(data=df_sector_30, y="Sector",
              order=sector_count_top.index).set_title("섹터별 빈도수")
```

```python
# 색상 표
print(plt.colormaps())
```

```python
# Blues : 빈도수 낮은게 진함, Blues_r : 빈도수 높은게 진함
plt.figure(figsize=(10,8))
sns.countplot(data=df_sector_30, y="Sector", palette="Blues_r",
              order=sector_count_top.index).set_title("섹터별 빈도수")
```

### Industry 빈도수
* 빈도수가 높은 top30개만!

```python
industry_count_top = df["Industry"].value_counts().head(30)
industry_count_top
```

```python
# sort_values() : 순차정렬
industry_count_top.sort_values().plot.barh(figsize=(10,8), title="산업별 빈도수")
```

```python
df_Industry_30 = df[df["Industry"].isin(industry_count_top.index)]
df_Industry_30
```

```python
# 색상 표
print(plt.colormaps())
```

```python
plt.figure(figsize=(10,8))
sns.countplot(data=df_Industry_30, y="Industry",palette="spring_r",
             order=industry_count_top.index).set_title("산업별 빈도수")
```
