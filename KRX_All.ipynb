### Describe() 해석
* freq : 최빈값(top에 표시) -> freq가 1이면 유일한 값!
* count : 빈도수
* unique : 유일한 값 개수(count와 같으면 모두 유일한 값(freq=1)
* first, last : 날짜 데이터의 처음과 끝
