## 종목코드 추출 함수
- 함수 만들기 : 종목 이름으로 종목 코드 추출하는 함수

```python
df_krx[["Symbol", "Name"]].head()
```

```python
# 종목코드(Symbol) 추출
df_krx.loc[df_krx["Name"] == "하이브", "Symbol"].tolist()[0]
```

```python

item_name = "하이브"
df_krx.loc[df_krx["Name"] == item_name, "Symbol"].tolist()[0]
```

```python
def item_code_by_item_name(item_name):
    """
    종목명을 받아 종목코드를 찾아 반환하는 함수
    """
    item_code = df_krx.loc[df_krx["Name"] == item_name, "Symbol"].tolist()
    
    # 예외처리 필요 : list index out of range(오타가 났을때 index error)
    if len(item_code) > 0 :
        return item_code[0]
    else :
        return False
```

## 종목명으로 일별시세 받아오는 함수

```python
def find_item_list(item_name, year=2020):
    """
    종목명으로 일별시세를 반환하는 함수
    item_code_by_item_name 사용
    """
    
    item_code = item_code_by_item_name(item_name)
    
    if item_code : 
        df_day = fdr.DataReader(item_code, str(year))
        return df_day
    else : 
        return False
```

```python
find_item_list("삼성전자", 2018)
```

```python
find_item_list("삼성")
```

## 종목별 일별시세 시각화

```python
stock_daily = find_item_list("하이브")
stock_daily
```
